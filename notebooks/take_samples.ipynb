```python
def process_piece(im):
    return cv2.GaussianBlur(im, (5, 5), 5, 5)
```

```python
base_path = './generated/'

def process_image(dirname, path):
    full = cv2.imread(path)
    full = cv2.cvtColor(full, cv2.COLOR_BGR2RGB)
    _, file = os.path.split(path)
    _, dirname = os.path.split(dirname)
    
    for n, piece in enumerate(map(process_piece, (full[512 * i: 512 * (i + 1), 512 * j: 512 * (j + 1)] for i in range(4) for j in range(4)))):
        imsave(os.path.join(base_path, '{}_{}_{}.jpg'.format(dirname, file, n)), piece)
    
#process_image(sample_dir, last_sample)
```

```python
from tqdm import tqdm_notebook as tqdm

def process_dir(dirname):
    files = []
    for f in tqdm(glob.glob(dirname + '/*'), desc=dirname):
        epoch = int(f.split('train_')[-1].split('_')[0])
        if epoch < 20:
            continue
        process_image(dirname, f)
```
