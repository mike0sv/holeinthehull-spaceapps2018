```python
from multiprocessing.dummy import Pool
```

```python
import glob
import json
import os
from tqdm import tqdm_notebook as tqdm
```

```python
def read_meta(path):
    with open(path, 'r', encoding='utf8') as f:
        return os.path.split(path)[-1].split('.')[0], json.load(f)
    
metas = dict(map(read_meta, glob.glob('hubble/meta/*')))
```

```python
import urllib.request

def download_from_meta(item):
    iid, meta = item
    try:
        file_meta = max((im for im in meta['image_files'] if im['file_url'].endswith('jpg')), key=lambda x: x['width'])
    except:
        return
    if not os.path.isfile('hubble/jpgs/{}.jpg'.format(iid)):
        print('Downloading', file_meta['file_url'])
        urllib.request.urlretrieve(file_meta['file_url'], 'hubble/jpgs/{}.jpg'.format(iid))
```

```python
with Pool(40) as pool:
    list(tqdm(pool.imap_unordered(download_from_meta, metas.items()), total=len(metas)))
```
