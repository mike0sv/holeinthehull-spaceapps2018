```python
import glob, markovify, json
```

```python
text_list = []

for file in glob.glob('../hubble/meta/*'):
    with open(file, 'r', encoding='utf8') as desc:
        try:
            text_list.append(json.load(desc)['description'])
        except:
            continue
```

```python
with open('corpus.txt', 'w', encoding='utf8') as file:
    file.write(' '.join(text_list))
```

```python
model = markovify.Text(' '.join(text_list))
```

```python
model.make_sentence(text_output = False, tries = 1000, init_state='Mars')
```

```python
#Generating texts

textgen.load('models/space.hdf5')

for temp in [0.2, 0.5, 0.65, 0.8, 1.0]:
    gen_text_list = []
    
    while len(gen_text_list) < 100:
        
        gen_texts = textgen.generate(10, temperature=temp, return_as_list=True)
        gen_texts = [text for text in gen_texts if text != '' and len(list(text)) > 150]
        
        gen_text_list.extend(gen_texts)
        print(len(gen_text_list))
    
    with open('gen_texts/space-temp{}.txt'.format(temp), 'w', encoding = 'utf8') as file:
        for text in gen_text_list:
            file.write("%s\n" % text)
```
